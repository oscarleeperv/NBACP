```python
import pandas as pd

#regression models
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
from sklearn.svm import NuSVR
```

```python
#read data to dataframe
df = pd.read_excel('data_ready.xlsx')
```

```python
#rookie season model to predict minutes per game
X = df[['COL_PTS_PERG', 'COL_FG_PCT']]
y = df['PRO_MIN_PERG']
mpg_model = NuSVR()
mpg_model.fit(X, y)
```

```python
#rookie season model to predict points per game
X = df[['COL_PTS_PERG']]
Y = df['PRO_PTS_PERG']
ppg_model = LinearRegression()
ppg_model.fit(X, y)
```

```python
#rookie season model to predict rebounds per game
X = df[['COL_REB_PERG', 'COL_FG_PCT']]
y = df['PRO_REB_PERG']
rpg_model = linear_model.Ridge(alpha=1)
rpg_model.fit(X, y)
```

```python
#rookie season model to predict assists per game
X = df[['COL_AST_PERG', 'COL_FG_PCT']]
y = df['PRO_AST_PERG']
apg_model = LinearRegression()
apg_model.fit(X, y)
```

```python
#rookie season model to predict field goal percentage
X = df[['COL_FG_PCT', 'COL_REB_PERG']]
y = df['PRO_FG_PCT']
fgpct_model = linear_model.Ridge(alpha=0.4)
fgpct_model.fit(X, y)
```

```python
#function puts all models together to predict a player's basic rookie season performance statistics
def pred_rookie_season(dict):
    mpg_input = pd.DataFrame({'COL_PTS_PERG': [dict['PPG']],'COL_FG_PCT': [dict['FG_PCT']]})
    mpg = mpg_model.predict(mpg_input)[0]
    ppg_input = pd.DataFrame({'COL_PTS_PERG': [dict['PPG']]})
    ppg = ppg_model.predict(ppg_input)[0]
    rpg_input = pd.DataFrame({'COL_REB_PERG': [dict['RPG']],'COL_FG_PCT': [dict['FG_PCT']]})
    rpg = rpg_model.predict(rpg_input)[0]
    apg_input = pd.DataFrame({'COL_AST_PERG': [dict['RPG']],'COL_FG_PCT': [dict['FG_PCT']]})
    apg = apg_model.predict(apg_input)[0]
    fgpct_input = pd.DataFrame({'COL_FG_PCT': [dict['FG_PCT']],'COL_REB_PERG': [dict['RPG']]})
    fgpct = fgpct_model.predict(fgpct_input)[0]

    print(dict['name'])
    print(f'MPG: {round(mpg, 1)}')
    print(f'PPG: {round(ppg, 1)}')
    print(f'RPG: {round(rpg, 1)}')
    print(f'APG: {round(apg, 1)}')
    print(f'FG%: {round(fgpct*100, 1)}%')
```

```python
#college basketball stats from players selected in the 2024 NBA draft
players = [{'name': 'Donovan Clingan', 
            'PPG': 13.0, 'RPG': 7.4,
            'APG': 1.5, 'FG_PCT': 0.639},
           {'name': 'Reed Sheppard', 
            'PPG': 12.5, 'RPG': 4.1,
            'APG': 4.5, 'FG_PCT': 0.536},
           {'name': 'Zach Edey', 
            'PPG': 25.2, 'RPG': 12.2,
            'APG': 2.0, 'FG_PCT': 0.623},
           {'name': 'Lebron James Jr', 
            'PPG': 4.8, 'RPG': 2.8,
            'APG': 2.1, 'FG_PCT': 0.366}]
```

```python
#function takes in Lebron James Jr's college basketball stats and predict's his rookie season
#stats in the 2024-25 regular season
pred_rookie_season(players[3])
```
